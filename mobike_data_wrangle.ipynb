## 清理

```python
df_clean = df.copy()
```

### 质量

#### - start_time数据类型错误
#### - end_time数据类型错误

##### 定义
使用to_datatime将start_time和end_time转为date_time格式

##### 代码

```python
df_clean.start_time = pd.to_datetime(df_clean.start_time)
df_clean.end_time = pd.to_datetime(df_clean.end_time)
```

##### 测试

```python
df_clean.info()
```

### 删除不需要的变量

```python
df_clean.drop(['track'], axis=1, inplace=True)
```

##### 测试

```python
df_clean.head()
```

## 数据整理
### 添加骑行时长变量duration

```python
#计算结束时间和起始时间的差值得到duration，并将其单位转换成分钟
df_clean['duration']= pd.to_datetime(df_clean.end_time)-pd.to_datetime(df_clean.start_time)
df_clean['duration']= df_clean.duration.apply(lambda x:round(x.total_seconds()/60))
```

```python
df_clean
```

## 保存数据

```python
save_file_name = 'mobike_shanghai_sample_clean.csv'
df_clean.to_csv(save_file_name, encoding='utf-8',index=False)
```
